```python
#Dropping null values: Data points without county have no lat,lon data 
df_city_shoot_merge_clean = df_cities_shoots_merged[df_cities_shoots_merged['County'].notnull()]
```

```python
df_city_shoot_merge_clean.isnull().sum()
```

4. Merging population with master dataset (mass shootings_gun sales_ cities)

```python
#Homogenizing county names
df_pop['County'] = df_pop['County'].str.replace(' County', '', regex=False).str.strip()
```

```python
df_pop
```

```python
#Melting population dataset
#Renaming columns on 
df_pop = df_pop.rename(columns = {
    'Unnamed:0': 'ID',
    'Pop2015': '2015',
    'Pop2016': '2016',
    'Pop2017': '2017',
    'Pop2018': '2018',
    'Pop2019': '2019',
    'Pop2020': '2020',
    'Pop2021': '2021',
    'Pop2022': '2022',
    'Pop2023': '2023',
    'Pop2024': '2024'
   })
```

```python
df_pop
```

```python
#Melting
year_columns = [str(y) for y in range(2015, 2025)] 
```

```python
df_pop_long = pd.melt(
    df_pop,
    id_vars=['State', 'County'],   # keep these
    value_vars=year_columns,       # only melt year columns
    var_name='Year',
    value_name='Population'
)

df_pop_long['Year'] = df_pop_long['Year'].astype(int)
```

```python
df_pop_long
```

```python
#Dropping and changing names in df_city_shoot_merge_clean
#Renaming columns on 
df_city_shoot_merge_clean = df_city_shoot_merge_clean.rename(columns = {
    'State_x': 'State'
   })
```

```python
#Homogenizing names
# For your incidents dataset
df_city_shoot_merge_clean['County'] = df_city_shoot_merge_clean['County'].str.strip().str.title()
df_city_shoot_merge_clean['State'] = df_city_shoot_merge_clean['State'].str.strip().str.title()
df_city_shoot_merge_clean['Year'] = df_city_shoot_merge_clean['Year'].astype(int)

# For your melted population dataset
df_pop_long['County'] = df_pop_long['County'].str.strip().str.title()
df_pop_long['State'] = df_pop_long['State'].str.strip().str.title()
df_pop_long['Year'] = df_pop_long['Year'].astype(int)
```

```python
#Check 
print(df_city_shoot_merge_clean.columns)
print(df_pop_long.columns)
```

```python
#Merging city_shoot_merge_clean with population
df_merged = pd.merge(
    df_city_shoot_merge_clean,
    df_pop_long,
    on=['State', 'County', 'Year'],
    how='left'
)
```

```python
df_merged
```

```python
null_columns = df_merged.columns[df_merged.isnull().any()].tolist()
print(null_columns) 
```

```python
df_merged.isnull().sum()
```
