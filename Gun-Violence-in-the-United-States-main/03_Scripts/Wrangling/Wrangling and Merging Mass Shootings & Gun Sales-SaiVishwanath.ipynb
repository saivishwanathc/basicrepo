3. Parsing and splitting date column in mass shootings dataset to match with other datasets

```python
# Create a new column with datetime format 
df_massshoot['Date'] = pd.to_datetime(df_massshoot['Incident_Full_Date'])

# Extract year and month number
df_massshoot['Year'] = df_massshoot['Date'].dt.year
df_massshoot['Month'] = df_massshoot['Date'].dt.month
```

```python
df_massshoot.head()
```

4. Dropping unnecessary columns

```python
#Dropping in MASS SHOOTINGS Unnamed: 0, Incident_Full_Date, Address 
ms_wrangled=df_massshoot.drop(columns=['Unnamed: 0', 'Incident_Full_Date', 'Address'])
```

```python
ms_wrangled
```

```python
#Dropping in GUN SALES columns in df_gunsales
gs_wrangled = df_gunsales.drop(columns=['Unnamed: 0'])
```

```python
gs_wrangled 
```

5. Rearranging column order in ms_wrangled

```python
# Define the new order
cols = ms_wrangled.columns.tolist()

# Columns to move
cols_to_move = ['Date', 'Year', 'Month']

# Position after 'Incident_ID'
insert_after = 'Incident_ID'
insert_index = cols.index(insert_after) + 1

# Remove the columns to move
for col in cols_to_move:
    cols.remove(col)

# Insert them back at the desired position
for i, col in enumerate(cols_to_move):
    cols.insert(insert_index + i, col)

# Reorder the DataFrame
ms_wrangled2 = ms_wrangled[cols]
```

```python
ms_wrangled2
```

6. Renaming columns 

```python
#Renaming in MASS SHOOTINGS
ms_wrangled3 = ms_wrangled2.rename(columns = {
    'City_County': 'City'
})
```

```python
ms_wrangled3
```

```python
#Renaming columns in GUN SALES dataset 
gs_wrangled2 = gs_wrangled.rename(columns={
    'year': 'Year',
    'month' : 'Month',
    'state' : 'State',
    'handgun_unadjusted' : 'Handguns_Sold',
    'long_gun_unadjusted' : 'Long_Guns_Sold',
    'total_unadjusted' : 'Total_Guns_Sold'			
})
```

```python
gs_wrangled2
```

7. Creating key columns for merge

```python
# Combine year, month, state to one column in gs_wrangled2 to connect with mass shootings data
gs_wrangled2['Year_Month_State'] = gs_wrangled2['Year'].astype(str) + '-' + gs_wrangled2['Month'].astype(str) + '-' + gs_wrangled2['State'].astype(str) 
```

```python
# Combine year, month, state to one column in ms_wrangled3 to connect with mass shootings data
ms_wrangled3['Year_Month_State'] = ms_wrangled3['Year'].astype(str) + '-' + ms_wrangled3['Month'].astype(str) + '-' + ms_wrangled3['State'].astype(str) 
```

8. Merge

```python
# Merge datasets: Mass shootings and Gun sales data 
ms_gs_merged = pd.merge(ms_wrangled3, gs_wrangled2, on='Year_Month_State', how='left')
```

```python
ms_gs_merged
```

```python
#Dropping unnecessary columns 
ms_gs_merged = ms_gs_merged.drop(columns=['Year_y', 'Month_y', 'State_y'])
```

```python
#Renaming columns
ms_gs_merged = ms_gs_merged.rename(columns={
    'Year_x': 'Year',
    'Month_x' : 'Month',
    'State_x' : 'State',
})			
```
