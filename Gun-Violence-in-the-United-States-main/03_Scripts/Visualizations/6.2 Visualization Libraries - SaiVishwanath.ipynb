#### 4. Scatterplots 

```python
# Create a scatterplot for the "Victims_Killed" and "Total_Guns_Sold" columns in seaborn

sns.lmplot(x = 'Total_Guns_Sold', y = 'Victims_Killed', data = sub)
plt.show()
```

Interpretation: Since none of the variables present significant correlation, this scatterplot presents the relationship between
two variables that were thought to be correlated as an initial hypothesis: Total guns sold versus the number of victims killed.
A thing to note is that further wrangling or geographic visualization may be needed to identify further trends.  

#### 5. Pair plots

```python
# Keep only the variables you want to use in the pair plot

sub_2 = sub[['Victims_Killed', 'Victims_Injured', 'Suspects_Killed', 
          'Suspects_Injured', 'Suspects_Arrested', 'Handguns_Sold', 'Long_Guns_Sold', 
          'Total_Guns_Sold']]
```

```python
# Create a pair plot 
g = sns.pairplot(sub_2)
plt.show()
```

It seems the number of guns does not influence the amount of victims, be it injured or killed. Most of them conglomerate around
the lower values of guns sold, no matter the kind of gun. Victim numbers are also spread out relatively even across the axis of guns sold, 
which might mean victims occur simply with the presence of a gun, no matter the number. It would be interesting to confirm this obvservation with further analysis, and likewise, to see it display on a geographic visualization to find out if there are any spatial patterns at large.

#### 6. Categorical plots

##### Creating a categorical plot based on mortality of mass shooting incidents

```python
sns.histplot(df['Victims_Killed'], bins = 30, kde = True)
plt.show()
```

```python
df.loc[df['Victims_Killed'] < 20, 'Mortality'] = 'Low'
df.loc[(df['Victims_Killed'] >= 20) & (df['Victims_Killed'] < 50), 'Mortality'] = 'Medium'
df.loc[df['Victims_Killed'] >= 50, 'Mortality'] = 'High'
df['Mortality'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style='ticks')
g = sns.catplot(x='Total_Guns_Sold', y='Victims_Killed', hue= 'Mortality', data=df)
plt.show()
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style='ticks')
g = sns.catplot(x='Suspects_Arrested', y='Victims_Killed', hue= 'Mortality', data=df)
plt.show()
```
