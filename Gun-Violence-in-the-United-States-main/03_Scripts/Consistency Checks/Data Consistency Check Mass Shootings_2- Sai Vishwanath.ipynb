3. Identifying irregularities: Mixed-type columns

```python
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df
[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```

```python
#Result: Address column has mixed type data
```

```python
#Changing data type to string
df['Address'] = df['Address'].astype('str') 
```

```python
#Verifying change
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df
[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```
