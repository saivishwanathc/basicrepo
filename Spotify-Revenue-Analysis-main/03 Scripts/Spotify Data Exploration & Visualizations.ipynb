## 3. Data Visualization

```python
# Create visualization of overall growth of total revenue and its components (premium revenue vs. ad revenue) over time to track trends and seasonality

# Create a line plot for revenue trends
plt.plot(df['Year'], df['Total Revenue'], label='Total Revenue')
plt.plot(df['Year'], df['Premium Revenue'], label='Premium Revenue')
plt.plot(df['Year'], df['Ad Revenue'], label='Ad Revenue')

plt.legend()
plt.title('Revenue Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Revenue')

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Revenue_Trends_Over_Time.png'))

plt.show()
```

```python
# Updating 'Date' column formatting
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
# Rcreate a line plot for revenue trends with new date format
plt.plot(df['Date'], df['Total Revenue'], label='Total Revenue')
plt.plot(df['Date'], df['Premium Revenue'], label='Premium Revenue')
plt.plot(df['Date'], df['Ad Revenue'], label='Ad Revenue')

plt.legend()
plt.title('Revenue Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Revenue_Trends_Over_Time_DATE.png'))

plt.show()
```

**Observations:**
- Ad revenue remains relatively steady until 2020, then dips before recovering
- Total Revenue and Premium Revenue have shown consistent growth over the years, with Premium Revenue increasing at a more steady and linear rate
- As Ad Revenue increases, Premium and Total Revenue also increase

```python
# Compare Total Revenue and Cost of Revenue to evaluate efficiency and profitability over time.
plt.stackplot(df['Date'], df['Cost of Revenue'], df['Total Revenue'], labels=['Cost of Revenue', 'Total Revenue'])
plt.legend(loc='upper left')
plt.title('Revenue and Cost Efficiency Over Time')

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Revenue_Cost_Efficiency_Over_Time.png'))

plt.show()
```

**Observations:**
- As the Cost of Revenue increases over time, Total Revenue also rises, suggesting a proportional relationship.
- Total Revenue is growing faster than Cost of Revenue, indicating improved operational efficiency.

```python
# Display growth in Total Monthly Active Users (MAUs), Premium MAUs, and Ad-Supported MAUs alongside Total Revenue to see the relationship between user engagement and revenue.
import matplotlib.dates as mdates

fig, ax1 = plt.subplots()

# Plot for Monthly Active Users (MAUs)
ax1.plot(df['Date'], df['Monthly Active Users'], label='Total MAUs', color='blue')
ax1.plot(df['Date'], df['Premium MAUs'], label='Premium MAUs', color='orange')
ax1.plot(df['Date'], df['Ad MAUs'], label='Ad-Supported MAUs', color='purple')

# Labels and Title for left axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Monthly Active Users (MAUs)', color='blue')
ax1.set_title('MAUs Growth and Revenue Over Time')

# Format x-axis to show Month and Year
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.xticks(rotation=45)

# Create the second y-axis for Total Revenue
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Total Revenue'], color='green', label='Total Revenue')

# Labels for the right axis
ax2.set_ylabel('Total Revenue (in millions)', color='green')

# Legends for both axes
ax1.legend(loc='upper left')  # Legend for MAUs
ax2.legend(loc='upper right')  # Legend for Revenue

plt.tight_layout()

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'MAUs_Growth_Revenue.png'))

plt.show()
```

**Observations:**
- As user engagement increases, total revenue increases

```python
# Show the contribution of Premium Revenue and Ad Revenue to Total Revenue for a given time period.

labels = ['Premium Revenue', 'Ad Revenue']
sizes = [df['Premium Revenue'].sum(), df['Ad Revenue'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Revenue Contribution by Premium vs. Ad')

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'revenue_contribution_premium_v_ad.png'))

plt.show()
```

```python
# Stacked bar chart for operating costs vs total revenue
# Ensure the DataFrame is sorted by date
df = df.sort_index(ascending=True)

# Stacked bar chart for operating costs vs total revenue
plt.figure(figsize=(10, 6))
df[['Cost of Revenue', 'Sales and Marketing Cost', 'Research and Development Cost']].plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8)
plt.plot(df.index, df['Total Revenue'], color='black', label='Total Revenue', linewidth=2)

plt.title('Operating Costs and Total Revenue Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cost / Revenue', fontsize=12)
plt.legend(['Total Revenue', 'Cost of Revenue', 'Sales & Marketing', 'R&D'], loc='upper left')

plt.tight_layout()

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'stacked_bar_costs_vs_revenue.png'))

plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

# Train an ARIMA model (assuming 'total_revenue' is the target variable)
train_size = int(len(df) * 0.8)
train, test = df['Total Revenue'][:train_size], df['Total Revenue'][train_size:]

model = ARIMA(train, order=(5, 1, 0))  # Adjust order based on data
model_fit = model.fit()

# Forecast for the next steps (the same length as test)
forecast = model_fit.forecast(steps=len(test))

# Plot actual vs forecasted revenue
plt.figure(figsize=(10, 6))
plt.plot(df.index[:train_size], train, label='Training Data', color='blue')
plt.plot(df.index[train_size:], test, label='Actual Revenue', color='green')
plt.plot(df.index[train_size:], forecast, label='Forecasted Revenue', color='red')

plt.title('Actual vs Forecasted Revenue', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.legend()

plt.tight_layout()

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'revenue_forecast_plot.png'))

plt.show()
```

```python

# Creating mock growth scenarios to project future revenue under different scenarios

# Scenario 1: Base Case (e.g., 5% revenue growth per period)
df['base_case_revenue'] = df['Total Revenue'] * (1 + 0.05)

# Scenario 2: Optimistic Scenario (e.g., 10% revenue growth per period)
df['optimistic_revenue'] = df['Total Revenue'] * (1 + 0.10)

# Scenario 3: Pessimistic Scenario (e.g., 2% revenue growth per period)
df['pessimistic_revenue'] = df['Total Revenue'] * (1 + 0.02)

# Plot the actual and scenario revenue projections
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Total Revenue'], label='Actual Revenue', color='blue')
plt.plot(df.index, df['base_case_revenue'], label='Base Case (5% Growth)', color='green', linestyle='--')
plt.plot(df.index, df['optimistic_revenue'], label='Optimistic (10% Growth)', color='orange', linestyle='--')
plt.plot(df.index, df['pessimistic_revenue'], label='Pessimistic (2% Growth)', color='red', linestyle='--')

# Formatting the plot
plt.title('Revenue Scenario Analysis', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.legend()
plt.tight_layout()

# Save the plot to the specified folder
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'revenue_scenario_analysis.png'))


# Show the plot
plt.show()
```
